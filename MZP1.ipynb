###Aufgabe 2
Für die $(4 \times 4)$-Matrizen $A$, $B$ sei folgendes bekannt
$$
\det (A^{-1}) = \frac{1}{2}, \det ((AB)^{-1}) = 7.
$$
Berechnen Sie, falls möglich (ansonsten schreiben Sie nicht möglich)

  * $\det B$,
  * $\det(A+B)$,
  * $\det (AB)$,
  * $\det (A^T B^{-1} B A^{-1})$,
  * $\det 2B$,
  * den Rang von $B$.

###Lösung 

#### B)

A durch A inverse berechnen

$
\det(A^{-1})=\large{\frac{1}{ \det(A)}} \Longrightarrow det(A) = 2
$


B inverse durch A berechnen

$\det (B^{-1})\cdot \det(A^{-1}) = 7 \Longrightarrow \det (B^{-1}) = 14$

B durch B inverse berechnen

$
\det(B)=\large{\frac{1}{\det(B^{-1})}} \Longrightarrow det(B) = \frac{1}{14}
$

#### A+B)
Wir kennen die Matrizen leider nicht und die Bedingung gilt nicht immer...
$$det(A+B) \neq det(A) + det(B)$$
nicht möglich

####AB)
**Produktsatz für Determinanten.**
$$
\det(A \cdot B) = \det(A) \cdot \det(B) = 2 \cdot \frac{1}{14} = \frac{1}{7}. 
$$

#### $\bf A^T B^{-1} B A^{-1} )$
$\det (A^T B^{-1} B A^{-1}) = \det (A^T A^{-1}) = \det(A^T) \cdot \det(A^{-1}) = \det(A) \cdot \large{\frac{1}{ \det(A)}} = 1$

####2B)
$\det(cB) = c^n*\det(B) = 2^4 \cdot \large\frac{1}{14} = \frac{8}{7}$

####rang B)
Bei einer linearen Abbildung $f$ ist der Rang als Dimension des Bildes $\mathrm{im}(f)$ dieser Abbildung definiert:
$$\mathrm{rang}(f) = \dim (\mathrm{im}(f))$$

Ist der Rang einer quadratischen Matrix gleich ihrer Zeilen- und Spaltenzahl, hat sie vollen Rang und wird reguläre Matrix genannt. Diese Eigenschaft lässt sich anhand ihrer Determinante feststellen. Eine Matrix hat genau dann vollen Rang, wenn ihre Determinante von null verschieden ist bzw. keiner ihrer Eigenwerte null ist.


Determinante von B ist nicht 0, dh. es gibt keine linear abhängige Spalten und somit der Rang ist gleich Anzahl n linear unabhängige Spalten
$$rang(B) = n = 4$$

###Aufgabe 3

Sei $f: \mathbb{R}^2 \rightarrow \mathbb{R}^2$ gegeben durch

###Lösung

$ \underbrace{\mathbb{R}^2}_\text{kanonische Basen} \qquad  \xrightarrow{A}  \qquad   \underbrace{\mathbb{R}^2}_\text{kanonische Basen} $

$\downarrow T \qquad \qquad \qquad  \qquad \qquad \qquad \uparrow Q$

2) Alte Technik (empfohlen)

$$Q=I^{-1}

```python
import numpy as np
x = np.mat([[1,-1],[1,1]])
y = np.mat([[5,1],[2,0]])
A = y*x.I
print(A)
T = np.mat(np.eye(2))
S = T.I
print(S)
Q = np.mat([[2,3],[1,1]])
P = Q.I
print(P)
B = P*A*S
print(B)
```
