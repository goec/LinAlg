####Serie 1-2

Sei $f$ die lineare Abbildung gegeben durch

$$ f \left( \begin{array}{c} 1 \\ 0 \\ 0 \end{array} \right) = \left( \begin{array}{c} 1 \\ 1 \end{array} \right), \quad f \left( \begin{array}{c} 0 \\ 1 \\ 0 \end{array} \right) = \left( \begin{array}{c} 3 \\ 1 \end{array} \right), \quad f \left( \begin{array}{c} 0 \\ 0 \\ 1 \end{array} \right) = \left( \begin{array}{c} 1 \\ 4 \end{array} \right)$$

Bestimmen Sie die Matrix $A$, die $f$ bezüglich der kanonischen Basis im $\mathbb{R}^3$ bzw. $\mathbb{R}^2$ darstellt.
Wie sieht die Matrix $A$ aus, wenn man im Bildraum $\mathbb{R}^2$ die Basis $\{(1,1)^T , (3,1)^T \}$ gewählt?


```python
from IPython.display import Image
Image(filename='basis.png')
```

###Lösung

$ \underbrace{\mathbb{R}^3}_\text{kanonische Basen} \qquad  \xrightarrow{A}  \qquad   \underbrace{\mathbb{R}^2}_\text{kanonische Basen} $

$\downarrow T \qquad \qquad \qquad  \qquad \qquad \qquad \uparrow Q$
<br>

$ \underbrace{\mathbb{R}^3}_\text{kanonische Basen} \qquad  \xrightarrow{A'}   \qquad \qquad   \underbrace{\mathbb{R}^2}_\text{Basen} $

$ \qquad \qquad \qquad \qquad  \qquad \left\{ \left( \begin{array}{c} 1 \\ 1 \end{array} \right),\left( \begin{array}{c} 3 \\ 1 \end{array} \right) \right \} $
<br> <br>
weil wir in $\mathbb{R}^3$ beide Seiten kanonische Basen haben:

$$T=I^{-1}I \quad \Longrightarrow T = I $$ $I$ ist die Einheitsmatrix, welche aus der Kanonische Basen besteht

weil wir in $\mathbb{R}^2$ nicht beide Seiten kanonische Basen haben, ziehen wir unsere Berechnung in Richtung der kanonischen Basen.

$$Q=I^{-1} 
\begin{pmatrix}
1 & 3  \\
1 & 1  \\
\end{pmatrix}  \quad \Longrightarrow Q = \begin{pmatrix}
1 & 3  \\
1 & 1  \\
\end{pmatrix} $$
Gesucht ist $$A'= P\,A\,S$$
Wir haben $A$(in Serie 1, Aufgabe 2), $T$ und $Q$ gefunden,dann können wir die anderen auch berechnen(siehe Abbildung):

$A = \begin{pmatrix}
1 & 3 & 1 \\
1 & 1 & 4 \\
\end{pmatrix} \qquad S = T^{-1} = \begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1 
\end{pmatrix}  \qquad P = Q^{-1} = \begin{pmatrix}
-0.5 & \,1.5 \\
\,0.5 & -0.5 
\end{pmatrix}$

$$ A'= P\,A\,S = \begin{pmatrix}
-0.5 & \,1.5 \\
\,0.5 & -0.5 
\end{pmatrix}\begin{pmatrix}
1 & 3 & 1 \\
1 & 1 & 4 \\
\end{pmatrix} \begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1 
\end{pmatrix} = \begin{pmatrix}
1 & 0 & 5.5 \\
0 & 1 & -1.5 \\
\end{pmatrix}$$




```python
A = np.mat([[1,3,1],[1,1,4]])
print(A)
T = np.mat(np.eye(3))
Q = np.mat([[1,3],[1,1]])
S = T.I
print(S)
P = Q.I
print(P)
A_new = P*A*S
print(A_new)
```

###Aufgabe 3

Betrachten Sie nochmals Aufgabe $4$ aus Serie $1$. Dort wurden zwei verschiedene

#### Serie 1, Aufgabe 4

Durch die Vorschrift $f(x) = Ax$ mit 
$$ A = 
\begin{pmatrix}
1 & 1 & 0 \\
2 & 1 & 1 \\
\end{pmatrix}
$$

ist eine lineare Abbildung von $\mathbb{R}^3$ nach $\mathbb{R}^2$ definiert(A stellt $f$ bezüglich der kanonischen
Basis dar). Bestimmen Sie Basen von $\mathbb{R}^3$ bzw. $\mathbb{R}^2$ , so dass man $f$ durch die Matrix

$$ B = 
\begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
\end{pmatrix}
$$
darstellen kann bzgl. diesen Basen.

###Lösung


$ \underbrace{\mathbb{R}^3}_\text{kanonische Basen} \qquad  \xrightarrow{A}  \qquad   \underbrace{\mathbb{R}^2}_\text{kanonische Basen} $

$\uparrow S \qquad \qquad \qquad  \qquad \qquad \qquad \uparrow Q$
<br>

$ \quad \underbrace{\mathbb{R}^3}_\text{Basen}   \qquad \quad \xrightarrow{B}   \qquad \qquad  \underbrace{\mathbb{R}^2}_\text{ Basen}   $

$ \left\{ \left( \begin{array}{c} 1 \\ 0 \\ 0 \end{array} \right),\left( \begin{array}{c} 0 \\ 1 \\ 0 \end{array} \right),\left( \begin{array}{c} -1 \\ 1 \\ 1 \end{array} \right) \right\}   \qquad   \left\{ \left( \begin{array}{c} 1 \\ 2 \end{array} \right),\left( \begin{array}{c} 1 \\ 1 \end{array} \right) \right\}  $

wir berechnen in $\mathbb{R}^2$ in Richtung kanonische Basen haben:

$Q = \begin{pmatrix}
1 & 1  \\
2 & 1  \\
\end{pmatrix}$

wir berechnen in $\mathbb{R}^3$ in Richtung kanonische Basen haben:

$S=\begin{pmatrix}
1 & 0 & -1 \\
0 & 1 & \,1 \\
0& 0 & \,1 
\end{pmatrix}$

Gesucht ist:

$$B= P\,A\,S$$

A ist gegeben, $S$ und $Q$ haben wir gefunden,dann können wir die andere Matrizen auch berechnen(siehe Abbildung):
<br>

$A = 
\begin{pmatrix}
1 & 1 & 0 \\
2 & 1 & 1 \\
\end{pmatrix}  \qquad S = \begin{pmatrix}
1 & 0 & \,1\\
0 & 1 & -1\\
0 & 0 & \,1
\end{pmatrix} \qquad P = Q^{-1} = \begin{pmatrix}
-1 & \,1  \\
\,2 & -1  \\
\end{pmatrix} $


$$ B= P\,A\,S = \begin{pmatrix}
-1 & \,1  \\
\,2 & -1  \\
\end{pmatrix} \begin{pmatrix}
1 & 1 & 0 \\
2 & 1 & 1 \\
\end{pmatrix} \begin{pmatrix}
1 & 0 & \,1\\
0 & 1 & -1\\
0 & 0 & \,1
\end{pmatrix} = \begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
\end{pmatrix}$$



```python
A = np.matrix([[1,1,0],[2,1,1]])
print(A)
S = np.mat([[1,0,-1],[0,1,1],[0,0,1]])
print(S)
Q = np.mat([[1,1],[2,1]])
P = Q.I
print(P)
B = P*A*S
print(B)
```
